```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams

import seaborn as sb
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sb.set_style('whitegrid')
```

**Defining PLOT color**

```python
x = range(1, 11)  # Now 10 values: 1 to 10 inclusive
y = [1, 2, 3, 4, 0, 5, 4, 3, 2, 1]

plt.bar(x, y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Bar Chart")
plt.show()
```

```python
x = range(1, 11)
y = [1, 2, 3, 4, 0, 5, 4, 3, 2, 1]
wide = [0.5, 0.5, 0.5, 0.9, 0.9, 0.9, 0.5, 0.5, 0.5, 0.5]  # now 10 widths
color = 'salmon'  # single color

plt.bar(x, y, width=wide, color=color, align='center')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Bar Chart with Varying Widths")
plt.show()
```

```python

x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

plt.plot(x, y1, linestyle='-', label='Solid Line')
plt.plot(x, y2, linestyle='--', label='Dashed Line')

plt.legend()
plt.title("Line Style Examples")
plt.show()
```

```python
data = {
    'A': [1, 3, 5, 7],
    'B': [2, 4, 6, 8],
    'C': [5, 3, 6, 2]
}
df = pd.DataFrame(data)

# Corrected color list
color_theme = ['darkgray', 'lightsalmon', 'powderblue']

# Plot
df.plot(color=color_theme)
plt.title("Line Plot with Custom Color Theme")
plt.show()
```

```python
z = [1,2,3,4,0,5]
plt.pie(z)
plt.show()
```

```python
color_theme = ['#A9A9A9','#FFA07A','#B0E0E6','#FFE4C4','#BDB76B']
plt.pie(z, colors=color_theme)
plt.show()
```

**customizing Line Styles**

```python
x1 = range(1,11)
y1 = [1, 2, 3, 4, 0, 5, 4, 3, 2, 1]
plt.plot(x,y)
plt.plot(x1,y1)
```

```python
plt.plot(x,y, ls = 'solid', lw=5)
plt.plot(x1,y1, ls='--', lw=10)
```

**setting plot markers**

```python
plt.plot(x,y, marker = '1', mew = 20)
plt.plot(x1,y1, marker= '+', mew=15)
```
