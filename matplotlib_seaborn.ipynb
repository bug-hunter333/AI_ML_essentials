```python
import matplotlib.pyplot as plt
import seaborn as sns

from pandas import DataFrame
```

```python
#we created the data set
data = {
    'names': ['steve', 'john', 'richard', 'sarah', 'randy', 'michael', 'julie'],
    'age': [20, 22, 20, 21, 24, 23, 22],
    'gender':['m', 'm', 'm', 'f', 'm', 'm', 'f'],
    'rank':['one', 'two', 'three', 'four', 'five', 'six', 'seven']
}

df = DataFrame(data)
df
```

```python
#matplolib bar chart

plt.bar(df['names'],df['age'])
plt.xlabel('names')
plt.ylabel('age')
plt.title('bar chart')

plt.show()
```

```python
#Geenrated using Seaborn

plot = sns.barplot(data=df, x='names', y='age')
plot.set_title('bar chart')
plot.set_xlabel('names')
plot.set_ylabel('age')

plt.show()
```

```python
#line plot

plt.plot(df['names'], df['age'])
plt.xlabel('names')
plt.ylabel('age')
plt.title('line chart')

plt.show()
```

```python
#line plt seaborn

plot = sns.lineplot(data=df, x='names',y='age')

plt.show()
```

```python
#Pie chart Matplotlib

plt.pie(df['age'], labels = df['names'])
plt.title("Age Comparison")
plt.show()
```

```python
#Pie chart SeaBorn

colors = sns.color_palette('pastel')[0:5]
plt.pie(df['age'], labels=df['names'], colors=colors)
plt.title("Age Comparison")
plt.show()
```
