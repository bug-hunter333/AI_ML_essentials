**VISUALIZING Time Series**

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams

import seaborn as sns
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sns.set_style('whitegrid')
```

**The Simplest Time series PLOT**

```python
address = '/sample_Data.csv'
# Option 1 (default, usually works for English text):
df = pd.read_csv(address, index_col='Order Date', parse_dates=True)

df.head()
```

```python
df['Order Quantity'].plot()
```

```python
# Example using fix Option A
df2 = df.sample(n=10, random_state=25)

plt.figure(figsize=(12, 6))
plt.xlabel('Order Date')
plt.ylabel('Order Quantity')
plt.title('Superstore Sales')

df2['Order Quantity'].plot()
plt.show()

```
